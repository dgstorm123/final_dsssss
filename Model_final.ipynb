**2-Data cleaning**

```python
# Fill missing values for less critical columns
df['old price'].fillna('0', inplace=True)
df['discount'].fillna('0%', inplace=True)
df['cpu generation'].fillna('Unknow', inplace=True)
df['gpu processor'].fillna('Apple', inplace=True)
df['rating'].fillna('0', inplace=True)
```

```python
df['new price'] = df['new price'].apply(lambda x: "{:,.0f} VND".format(float(x.replace('₹', '').replace(',', '')) * 304.75))
df['old price'] = df['old price'].apply(lambda x: "{:,.0f} VND".format(float(x.replace('₹', '').replace(',', '')) * 304.75))
```

```python
# Định dạng lại cột new price vnd để dễ nhìn hơn theo kiểu tiền tệ Việt Nam

df['new price vnd'] = df['new price']
df['old price vnd'] = df['old price']
```

```python
# xử lý new price và old price về float để cho vào mô hình
df['new price'] = df['new price'].str.replace('VND', '').str.replace(',', '')
df['new price'] = df['new price'].astype(float)

df['old price'] = df['old price'].str.replace('VND', '').str.replace(',', '')
df['old price'] = df['old price'].astype(float)
```

```python
# Trích xuất kích thước màn hình
df['screen size'] = df['screen size'].str.extract(r'\(([^)]+)\)')
```

```python
# Nhóm và đếm các giá trị duy nhất trong cột 'screen'
df['screen resolution'] = df['screen resolution'].str.rsplit(n=1).str[0]
screen_counts = df['screen resolution'].value_counts()

# Hiển thị kết quả
print(screen_counts)
```

```python
#làm sạch cột screen resolution
df['screen resolution'] = df['screen resolution'].str.replace('$$Pixels', '') \
                                                 .str.replace('X', 'x') \
                                                 .str.replace('1920x', '1920 x ') \
                                                 .str.replace('Pixels', '') \
                                                 .str.replace('pixel', '') \
                                                 .str.replace('x  1080', 'x 1080')
```

```python
# Nhóm và đếm các giá trị duy nhất trong cột 'screen'
screen_counts = df['screen resolution'].value_counts()

print(screen_counts)
```

```python
def convert_screen_resolution(resolution):
    # List of resolutions considered as '2k'
    if resolution == '1920 x 1080':
        return 'full hd'
    elif resolution == '3840 x 2160':
        return '4k'
    else:
        return '2k'

# Áp dụng hàm chuyển đổi cho cột 'screen '
df['screen resolution'] = df['screen resolution'].apply(convert_screen_resolution)
```

```python
# Loại bỏ phần trăm và chữ "off" trong cột Discount
def clean_discount(discount):
    if isinstance(discount, str):
        discount = discount.replace('%', '').replace('off', '').strip()
        return float(discount) / 100
    return discount

df['discount'] = df['discount'].apply(clean_discount)
```

```python

# Bỏ dấu & và ; trong cột Suitable for
df['suitable for'] = df['suitable for'].str.replace('&', 'and').str.replace(';', ',')

# Loại bỏ "Gen" trong cột Cpu generation và xử lý giá trị thiếu
df['cpu generation'] = df['cpu generation'].str.replace('Gen', '').str.strip()
df['cpu generation'].fillna('Unknown', inplace=True)

# Loại bỏ "GB" trong cột Ssd capacity và Ram capacity
df['ssd capacity'] = df['ssd capacity'].str.replace('GB', '').str.strip()
df['ssd capacity'] = df['ssd capacity'].str.replace('TB', '').str.strip()
df['ram capacity'] = df['ram capacity'].str.replace('GB', '').str.strip()

# Biến đổi giá trị NA trong cột GPU processor thành "apple"
df['gpu processor'].fillna('apple', inplace=True)


```

```python
def convert_os_system(system):

    if 'Windows 11 Home' in system or 'Windows 10 Home' in system:
        return 'Windows'
    elif 'macOS Ventura'or 'Mac OS Mojave' in system or 'Mac OS Monterey' and 'macOS Sonoma' or 'Mac OS Big Sur' in system:
        return 'Mac OS'
    else:
        return 'Ubuntu/linux'

# Áp dụng hàm chuyển đổi cho cột 'screen'
df['os system'] = df['os system'].apply(convert_os_system)
```

```python
# Sử dụng biểu thức chính quy để trích xuất phần cần thiết từ cột 'name'
df['name'] = df['name'].str.extract(r'^([^-\(]+(?:\([^)]*\))?[^-]*)')
```

```python
#Tạo cột brand
df['brand'] = df['name'].apply(lambda x: x.split()[0] if isinstance(x, str) else '')
```

```python
# Drop the 'description' column
df = df.drop(columns=['links'])
```

```python
import pandas as pd
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Assuming df is your DataFrame
# df = pd.read_csv('your_laptop_data.csv')

# Fill and clean the 'description' column
df['description'] = (
    df['series'].fillna('') + ' ' +
    df['os system'].fillna('') + ' ' +
    df['suitable for'].fillna('')
)
df['description'] = df['description'].str.strip()
df = df[df['description'].apply(lambda x: len(x) > 0)]

# Calculate TF-IDF
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['description'])

# Calculate cosine similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Function to get recommendations
def get_recommendations(suitable_for, os_system, budget, df, cosine_sim, top_n=10):
    # Filter data based on user criteria
    filtered_data = df[
        df['suitable for'].str.contains(suitable_for, case=False, na=False) &
        df['os system'].str.contains(os_system, case=False, na=False) &
        (df['new price'] <= budget)
    ]

    # If no products match, return a message
    if filtered_data.empty:
        return "Không có sản phẩm nào phù hợp với yêu cầu của bạn."

    # Get indices of the filtered data
    filtered_indices = filtered_data.index.tolist()

    # Map the filtered indices to positions in the original DataFrame
    index_mapping = {index: pos for pos, index in enumerate(df.index)}

    # Get the positions of the filtered indices
    filtered_positions = [index_mapping[idx] for idx in filtered_indices]

    # Calculate similarity scores for the filtered data
    sim_scores = cosine_similarity(tfidf_matrix[filtered_positions], tfidf_matrix)

    # Get mean similarity scores
    mean_sim_scores = sim_scores.mean(axis=0)

    # Sort by similarity scores
    similarity_scores = list(enumerate(mean_sim_scores))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    # Get top N similar products
    top_products = [i[0] for i in similarity_scores[:top_n]]
    top_scores = [i[1] for i in similarity_scores[:top_n]]

    # Return top N recommended products with scores
    recommended_products = df.iloc[top_products][['name', 'description']].copy()
    recommended_products['score'] = top_scores

    return recommended_products

# Function to display recommendations
def show_recommendations(b):
    clear_output()
    display(ui)
    suitable_for = suitable_for_dropdown.value
    os_system = os_system_dropdown.value
    budget = float(budget_input.value)

    recommendations = get_recommendations(suitable_for, os_system, budget, df, cosine_sim, top_n=10)

    if isinstance(recommendations, str):
        result_label.value = recommendations
    else:
        result_list.clear_output()
        with result_list:
            display(recommendations)

# Create widgets
suitable_for_dropdown = widgets.Dropdown(
    options=df['suitable for'].unique(),
    description='Nhu cầu:'
)

os_system_dropdown = widgets.Dropdown(
    options=df['os system'].unique(),
    description='Sở thích:'
)

budget_input = widgets.FloatText(
    description='Mức tài chính:'
)

recommend_button = widgets.Button(
    description='Gợi ý'
)

result_label = widgets.Label()
result_list = widgets.Output()

recommend_button.on_click(show_recommendations)

# Create UI
ui = widgets.VBox([
    suitable_for_dropdown,
    os_system_dropdown,
    budget_input,
    recommend_button,
    result_label,
    result_list
])

display(ui)
```
